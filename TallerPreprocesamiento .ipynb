### valores perdidos
Vamos a eliminar los valores faltantes, ya que tenemos suficientes datos y los json estaban vacíos

```python
df = df.dropna()
```

```python
num_cols = [
    'bpm', 'energy', 'danceability_ll', 'loudness', 'duration_ms',
    'danceability_prob', 'gender_prob', 'genre_dortmund_prob', 'genre_electronic_prob',
    'genre_rosamerica_prob', 'genre_tzanetakis_prob', 'ismir04_rhythm_prob',
    'mood_acoustic_prob', 'mood_aggressive_prob', 'mood_electronic_prob',
    'mood_happy_prob', 'mood_party_prob', 'mood_relaxed_prob',
    'mood_sad_prob', 'moods_mirex_prob', 'timbre_prob', 'tonal_atonal_prob',
    'voice_instrumental_prob'
]
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
df.info()
```

```python
## converti a dummies
for col in categorical_cols:
    if df[col].nunique()==2 and col in df.columns:
        dummies = pd.get_dummies(df[col], prefix=col)
        df = pd.concat([df, dummies], axis=1)
        df.drop(col, axis=1, inplace=True)
```

```python
#elimino las columnas que serían redundantes
dummies_to_drop = [
    "danceability_value_not_danceable",
    "gender_value_male",
    "mood_acoustic_value_not_acoustic",
    "mood_aggressive_value_not_aggressive",
    "mood_electronic_value_not_electronic",
    "mood_happy_value_not_happy",
    "mood_party_value_not_party",
    "mood_relaxed_value_not_relaxed",
    "mood_sad_value_not_sad",
    "timbre_value_dark",
    "tonal_atonal_value_atonal",
    "voice_instrumental_value_instrumental"
]
df = df.drop(columns=dummies_to_drop)
```

estos valores están en aprox el 50% de los valores, por lo que decidimos eliminar los valores, tenemos la probabilidad de que esté el género así que consideramos no hay mayor pérdida de info

```python
df["playlist_name"].fillna("unknown")
df["mbid"].fillna("unknown")
```
